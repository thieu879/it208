```python
# Bt1
def calculate_difference(num1, num2):
    difference = num1 - num2
    print("Hiệu của", num1, "-", num2, "=", difference)

def calculate_product(num1, num2):
    return num1 * num2

number1 = int(input("Nhập số thứ nhất: "))
number2 = int(input("Nhập số thứ hai: "))

calculate_difference(number1, number2)

product = calculate_product(number1, number2)
print("Tích của", number1, "*", number2, "=", product)
```

```python
# Bt2
def is_prime(number):
    if number <= 1:
        return False

    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False

    return True

num = int(input("Enter a number: "))

if is_prime(num):
    print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

```python
# Bt3
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

number = int(input("Nhập một số nguyên dương: "))
print("Giai thừa của", number, "là:", factorial(number))
```

```python
# Bt4
def circle_area(radius):
    pi = 3.14159
    area = pi * radius * radius
    return area

r = float(input("Nhập bán kính hình tròn: "))
print("Diện tích hình tròn là:", circle_area(r))
```

```python
# Bt5
def is_leap_year(year):
    if (year % 400 == 0):
        return True
    elif (year % 100 == 0):
        return False
    elif (year % 4 == 0):
        return True
    else:
        return False

y = int(input("Nhập năm: "))

if is_leap_year(y):
    print(y, "là năm nhuận")
else:
    print(y, "không phải là năm nhuận")
```

```python
# Bt6
def sum_of_digits(n):
    total = 0
    n = abs(n)
    while n > 0:
        total += n % 10 
        n //= 10
    return total

number = int(input("Nhập một số nguyên: "))
print("Tổng các chữ số là:", sum_of_digits(number))
```

```python
# Bt7
def handleReverseTheNumber(n):
    reversed_num = 0
    negative = False

    if n < 0:
        negative = True
        n = abs(n)

    while n > 0:
        digit = n % 10
        reversed_num = reversed_num * 10 + digit
        n //= 10

    if negative:
        reversed_num = -reversed_num

    return reversed_num

number = int(input("Nhập một số nguyên: "))
print("Số đảo ngược là:", handleReverseTheNumber(number))
```

```python
# Bt8
def is_perfect_number(n):
    if n < 1:
        return False
    
    sum_divisors = 0
    for i in range(1, n):
        if n % i == 0:
            sum_divisors += i
    
    return sum_divisors == n

number = int(input("Nhập một số nguyên: "))

if is_perfect_number(number):
    print(number, "là số hoàn hảo")
else:
    print(number, "không phải là số hoàn hảo")
```
