```python
# bt7
products = {
    "A": 120,
    "B": 80,
    "C": 300,
    "D": 50
}

new_products = {}
for k, v in products.items():
    if v >= 100:
        new_products[k] = v
    else:
        pass

products = new_products
print("Sản phẩm còn lại sau khi xóa:", products)

products["E"] = 150

print("Danh sách sản phẩm sau khi thêm:", products)

print("Số lượng sản phẩm hiện tại:", len(products))
```

```python
# Bt8
products = {
    "A": 120,
    "B": 80,
    "C": 300,
    "D": 50
}
if "A" in products:
    print("Giá sản phẩm A:", products["A"])
else:
    print("Sản phẩm A đã bị xóa")

sorted_products = dict(sorted(products.items(), key=lambda item: item[1]))
print("Danh sách sản phẩm sau khi sắp xếp theo giá:", sorted_products)

total_price = sum(products.values())
print("Tổng giá trị của tất cả sản phẩm:", total_price)

products = {k: v for k, v in products.items() if v <= 200}
print("Danh sách sản phẩm sau khi xóa giá > 200:", products)
```
