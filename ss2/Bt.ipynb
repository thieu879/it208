```python
# Bt4.py
n = int(input("nhập số: "))
sum = 0
if n > 0:
    for i in range(1, n + 1):
        sum += i
    print("sum =", sum)
else:
    print("vui lòng nhập số dương")
```

```python
# Bt5.py
a = int(input("Nhập số kiểm tra đối xứng: "))
original = a
reversed = 0
while a > 0:
    digit = a % 10
    reversed = reversed * 10 + digit
    a = a // 10
if original == reversed:
    print("là số đối xứng")
else:
    print("không phải là số đối xứng")
```

```python
# Bt6.py
n = int(input("nhập số: "))
if n > 0:
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    print("giai thừa =", factorial)
else:
    print("vui lòng nhập số dương")
```
